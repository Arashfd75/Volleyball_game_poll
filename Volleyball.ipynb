```python
class Player():

  def __init__(self, name, skill, presence):
    self.name = name
    self.skill = skill
    self.presence = presence
    self.points = 0


  
```

```python
class Team():

  def __init__(self, players):
    self.players = players
    self.number_of_games = 0
    self.points_played = 0
    self.inside = 0

  def in_players(self):
    output = []
    for i in range(len(self.players)):
      if self.players[i].presence == True:
        # print("i:",i)
        output += [self.players[i]]
    return output
  
  def get_ins(self):
    self.subs_in = [player for player in self.players if player.presence == True]
    self.subs_out = [player for player in self.players if player.presence == False]

  def substitute(self):
    self.subs_in[0].presence = False
    self.subs_out[0].presence = True
    # for out in self.subs_out:
    #   print(out.name) 
    # print("goes out:",self.subs_in[0].name)
    # print("goes in:",self.subs_out[0].name)
    # for out in self.subs_out:
    #   print(out.name) 
    self.subs_in[0], self.subs_out[0] = self.subs_out[0], self.subs_in[0]
    subs_in_deq = deque(self.subs_in)
    subs_out_deq = deque(self.subs_out)
    subs_in_deq.rotate(1)
    subs_out_deq.rotate(1)
    self.subs_in = list(subs_in_deq)
    self.subs_out = list(subs_out_deq)
    # print("000:", self.subs_out[0].name)
    # for out in self.subs_out:
    #   print(out.name) 

```

```python
def run_game(team1, team2):
  team1.inside = 1
  team2.inside = 1
  team1.number_of_games += 1
  team2.number_of_games += 1
  team1.points_played = round(random.normalvariate(11,2))
  team2.points_played = round(random.normalvariate(11,2))
  if team1.points_played > team2.points_played:
    team1.points_played = 15
  else:
    team2.points_played = 15
  return team1.points_played, team2.points_played
```

```python
def update_player_stats(t1,t2,result):
  total_point = result[1] + result[0]
  for i in range(total_point):
    for team in [t1, t2]:
      players_inside = team.in_players()
      # print(players_inside)
      # print("\n")
      for player in players_inside:
        player.points += 1
      if (i+1) % 6 == 0:
        
        team.substitute()    
  # print(t1.players[0].points)    

```
